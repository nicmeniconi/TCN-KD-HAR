```python
csv_path = os.path.join(dpath, 'S40', 'S40_A02_T01.sto')

# Open the file and skip the header (everything before 'endheader')
with open(csv_path, 'r') as file:
    # Read all lines
    lines = file.readlines()

# Find the line where the data starts (after 'endheader')
for i, line in enumerate(lines):
    if 'endheader' in line:
        start_idx = i + 1  # Data starts after the 'endheader'
        break

# Now read the data from the relevant lines and load it into a pandas DataFrame
from io import StringIO

# Join the data lines into a single string for pandas to read
data = ''.join(lines[start_idx:])

# Use StringIO to treat the string as a file and read it with pandas
df_sto02 = pd.read_csv(StringIO(data), sep='\s+')

# Display the DataFrame
print(f'Length: {len(df_sto02)}')
df_sto02.head()
```

```python
csv_path = os.path.join(dpath, 'S40', 'ik_S40_A01_T01_orientationErrors.sto')

data = []
columns = []
metadata_section = True  # Flag to skip metadata

with open(csv_path, 'r') as file:
    for line in file:
        line = line.strip()
        # Skip metadata section
        if metadata_section:
            if "endheader" in line:  # When the metadata ends
                metadata_section = False  # Switch to data reading mode
            continue
        
        # If we reach here, we're in the data section
        if line.startswith("time"):
            columns = line.split()  # Set column names when 'time' line appears
        else:
            data.append(line.split())  # Append numeric data

# Convert the data to a DataFrame
df = pd.DataFrame(data, columns=columns)

# Convert the values to appropriate data types
df_oerr02 = df.apply(pd.to_numeric)

# Display the first few rows
df_oerr02.head()
```
